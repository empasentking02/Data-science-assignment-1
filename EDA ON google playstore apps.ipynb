```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the URL
url = "https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv"
df = pd.read_csv(url)

# Display the first few rows of the DataFrame to understand its structure
print(df.head())

# Clean the dataset: Drop rows with missing values in the 'Category' column
df = df.dropna(subset=['Category'])

# Group by 'Category' and count the number of apps in each category
category_counts = df['Category'].value_counts()

# Select the top 10 categories
top_10_categories = category_counts.head(10)

# Plotting the top 10 app categories
plt.figure(figsize=(15, 6))
sns.barplot(x=top_10_categories.index, y=top_10_categories.values, palette='hls')

# Add plot title and rotate x-ticks for readability
plt.title('Top 10 App Categories')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.xticks(rotation=90)

# Show the plot
plt.show()






```
